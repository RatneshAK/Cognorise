## Exploratory Data Analysis On Customer Personality Analysis
Customer Personality Analysis is a detailed analysis of a company’s ideal customers. It helps a business to better understand its customers and makes it easier for them to modify products according to the specific needs, behaviors and concerns of different types of customers.

Customer personality analysis helps a business to modify its product based on its target customers from different types of customer segments. For example, instead of spending money to market a new product to every customer in the company’s database, a company can analyze which customer segment is most likely to buy the product and then market the product only on that particular segment.

```python
# Plot histogram for Year_Birth
plt.figure(figsize=(10, 6))
plt.hist(df['Year_Birth'], bins=range(1950, 2001, 5), color='skyblue', edgecolor='black')
plt.title('Distribution of Birth Years')
plt.xlabel('Year of Birth')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
# Bar plot for Education
plt.figure(figsize=(8, 5))
df['Education'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Distribution of Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
# Pie chart for Marital_Status
plt.figure(figsize=(8, 8))
df['Marital_Status'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral', 'lightgreen', 'lightsalmon', 'lightyellow'])
plt.title('Distribution of Marital Status')
plt.ylabel('')
plt.show()
```

```python
# Scatter plot for Income vs. Amount Spent On Wines
plt.figure(figsize=(10, 6))
plt.scatter(df['Income'], df['MntWines'], color='orange', alpha=0.7)
plt.title('Income vs. Amount Spent On Wines')
plt.xlabel('Income')
plt.ylabel('Amount Spent On Wines')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set_style("whitegrid")

# Create subplots
fig, axes = plt.subplots(3, 2, figsize=(15, 15))

# Plot histograms
sns.histplot(df['Year_Birth'], ax=axes[0, 0], bins=range(1950, 2001, 5), color='skyblue', edgecolor='black')
axes[0, 0].set_title('Distribution of Birth Years')

sns.histplot(df['Income'], ax=axes[0, 1], color='lightgreen', edgecolor='black')
axes[0, 1].set_title('Distribution of Income')

# Plot bar plots
sns.countplot(x='Education', data=df, ax=axes[1, 0], palette='pastel')
axes[1, 0].set_title('Distribution of Education Level')
axes[1, 0].tick_params(axis='x', rotation=45)

sns.countplot(x='Marital_Status', data=df, ax=axes[1, 1], palette='pastel')
axes[1, 1].set_title('Distribution of Marital Status')
axes[1, 1].tick_params(axis='x', rotation=45)

# Plot scatter plot
sns.scatterplot(x='Income', y='MntWines', data=df, ax=axes[2, 0], color='orange')
axes[2, 0].set_title('Income vs. Amount Spent On Wines')

# Plot pie chart
marital_status_counts = df['Marital_Status'].value_counts()
axes[2, 1].pie(marital_status_counts, labels=marital_status_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
axes[2, 1].set_title('Distribution of Marital Status')

# Adjust layout
plt.tight_layout()
plt.show()
```

### Conclusion

#### Birth Year Distribution:
- The distribution of birth years among customers is fairly uniform, with a slight increase in the number of customers born in the late 1970s to early 1980s.

#### Income Distribution:
- Income distribution among customers varies, with a majority falling within the middle-income range.
- There are fewer customers with high incomes compared to those with moderate incomes.

#### Education Level:
- The majority of customers have attained a graduate or undergraduate education level.
- A smaller proportion of customers have completed high school or postgraduate education.

#### Marital Status:
- The most common marital status among customers is married, followed by single.
- There is a smaller proportion of customers who are divorced or widowed.

#### Income vs. Amount Spent on Wines:
- There seems to be a positive correlation between income and the amount spent on wines, indicating that customers with higher incomes tend to spend more on wines.

#### Summary:
- The customer base consists of individuals with diverse characteristics in terms of birth year, income, education level, and marital status.
- Income appears to influence spending behavior, particularly in the case of wine purchases.
- Understanding these demographics and their purchasing patterns can inform targeted marketing strategies to better serve different customer segments.
